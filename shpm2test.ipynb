```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import CenteredNorm

from shpm2 import SHPM

from lds import lds
```

```python
n = 20
```

```python
shpm = SHPM(n)
```

```python
p = shpm.random_point()
p_ = shpm.mutate(p, mag=0.5, verbose=True)
```

```python
A = shpm.assemble(p)
A_ = shpm.assemble(p_)
```

# Compare Matrices

```python
plt.imshow(A, cmap="bwr", norm=CenteredNorm())
plt.colorbar()
```

```python
plt.imshow(A_, cmap="bwr", norm=CenteredNorm())
plt.colorbar()
```

```python
plt.imshow(A_ - A, cmap="bwr", norm=CenteredNorm())
plt.colorbar()
```

# Compare Eigenvalues

```python
eigs = np.linalg.eigvals(A)
eigs_ = np.linalg.eigvals(A_)
```

```python
plt.scatter(np.real(eigs), np.imag(eigs))
plt.vlines([0.0], colors="black", linestyles="--", ymin=-5.0, ymax=5.0)
plt.xlim(np.min(np.real(eigs)), 1.0)
plt.ylim(np.min(np.imag(eigs)), -np.min(np.imag(eigs)))
```

```python
plt.scatter(np.real(eigs_), np.imag(eigs_))
plt.vlines([0.0], colors="black", linestyles="--", ymin=-5.0, ymax=5.0)
plt.xlim(np.min(np.real(eigs)), 1.0)
plt.ylim(np.min(np.imag(eigs_)), -np.min(np.imag(eigs_)))
```

# Compare LDS behaviour


```python
x_0 = np.zeros((n,))
x_0[0] = 1.0
```

```python
xs = lds(A, x_0, dt=0.01, iters=400)
xs_ = lds(A_, x_0, dt=0.01, iters=400)
```

```python
plt.plot(xs)
```

```python
plt.plot(xs_)
```
