```python
import numpy as np
import pymanopt
import matplotlib.pyplot as plt

import matplotlib.cm as cm
import matplotlib.colors as mcolors
```

```python
n = 32
```

```python
grass = pymanopt.manifolds.Grassmann(2, 1)
```

```python
ls = [grass.random_point() for i in range(100)]
```

```python
cs = [grass.dist(l, ls[0]) for l in ls[1:]]
cn = mcolors.Normalize(vmin=min(cs), vmax=max(cs))
cmap = plt.colormaps.get_cmap("bwr")

plt.scatter([l[0] for l in ls[1:]], [l[1] for l in ls[1:]], c=[cmap(cn(val)) for val in cs])
plt.scatter([ls[0][0]], [ls[0][1]], c='g')
```

```python
stief = pymanopt.manifolds.Stiefel(n, n)
```

```python
Q_1 = stief.random_point()
Q_2 = stief.random_point()
```

```python
Q_1_ = stief.exp(Q_1, stief.random_tangent_vector(Q_1) * 0.1)
```

```python
plt.imshow(Q_1 - Q_1_, cmap="bwr")
plt.colorbar()
```
