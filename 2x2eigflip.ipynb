```python
import numpy as np
from lds import lds
import matplotlib.pyplot as plt
```

```python
alpha = -1.0
beta = -1.0
gamma = 2.0
```

```python
A_1 = np.array([[alpha, beta], [gamma, alpha]])
A_2 = np.array([[alpha, -beta], [-gamma, alpha]])
```

```python
x_0 = np.array([1, 0])
```

```python
xs_0 = lds(A_1, x_0, beta=0.0, iters=100)
xs_1 = lds(A_2, x_0, beta=0.0, iters=100)
```

```python
print(np.linalg.eigh(A_1))
print(np.linalg.eigh(A_2))
```

```python
plt.plot(xs_0)
```

```python
plt.plot(xs_1)
```
