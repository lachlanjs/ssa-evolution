```python
import numpy as np
import scipy as scp
from scipy.linalg import schur
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
n=6
```

```python
A = np.random.normal(0.0, 1.0, (n, n))
```

```python
T, Q = scp.linalg.schur(A, output="real")
```

```python
np.linalg.eigvals(A)
```

```python
sb.heatmap(T, annot=True)
```

```python
from scipy.linalg import rsf2csf

T, Q = schur(A, output='real')
T_complex, _ = rsf2csf(T, Q)

print("Eigenvalues from T:")
print(np.diag(T_complex))  # should match eigvals
```

```python
import matplotlib.pyplot as plt

plt.imshow(np.abs(T) > 1e-10, cmap='gray')
plt.title("Nonzero structure of T (real Schur)")
plt.show()
```
